Write a Python program to display calendar?

```python
l=['JAN','FEB','MAR','APRIL','MAY','JUNE','JULY','AUG','SEP','OCT','NOV','DEC']
for i in range(len(l)):
    if i==1:
        print(l[1])
        for j in range(1,29):
            print(j)
        print("----------")
    elif i%2==0:
        print(l[i])
        for j in range(1,32):
            print(j)
        print("----------")
    else:
        print(l[i])
        for j in range(1,31):
            print(j)
        print("----------")
            
```

Write a Python program to solve quadratic equation?

```python
import math
a=int(input("enter the coefficient of x^2- "))
b=int(input("enter the coefficient of x^1- "))
c=int(input("enter the coefficient of x^0- "))
d=b*b-4*a*c
if d<0:
    print("Imaginary roots , not possible to find")
elif d>=0:
    x1=(-b+math.sqrt(d))/(2*a)
    x2=(-b-math.sqrt(d))/(2*a)
    print(x1,x2)
    
```
